# 3.4   Regular Expressions for Detecting Word Patterns

```python
nltk.download('words')
```

```python
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
[w for w in wordlist if re.search('ed$', w)]
```

```python
# returns 8 letter words with 3rd letter j, 6th letter t. 
[w for w in wordlist if re.search('^..j..t..$', w)]
```

```python
[w for w in wordlist if re.search('^..j..tion$', w)]
```

```python
[w for w in wordlist if re.search('^..j..ti?on$', w)]
```

```python
#6284 -- numbers from a phone dial pad
[w for w in wordlist if re.search('^[mno][abc][tuv][ghi]$', w)]
```

```python
[w for w in wordlist if re.search('^[g-o]+$', w)]
```

```python
chat_words = sorted(set(w for w in nltk.corpus.nps_chat.words()))
```

```python
# '+' enables strings with 1 or more of that elememt
[w 
 for w in chat_words 
 if re.search('^m+i+n+e+$', w)]
```

```python
# '*' enables strings with 0 or more of that elememt
[w 
 for w in chat_words 
 if re.search('^m*i*n*e*$', w)]
```

```python
# '$' at the end signifies that that is the end of the sring.
# removing the $ sign leaves the string open ended. 
# -- mines is allowed.
[w 
 for w in chat_words 
 if re.search('^m+i+n+e+', w)]
```
